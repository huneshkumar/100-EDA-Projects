```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/Books_Data_Clean.csv')
df
```

```python
## show 1st five roles

df.head()
```

```python
df.info()
```

```python
# we have publishing year -560.000000

#fixing it 
```

```python
df=df[df['Publishing Year']>1900]

df.describe()
```

```python
# now rome rows that dont have book name 

df.dropna( subset='Book Name', inplace=True)
```

```python
#check duplicates

df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist( df['Publishing Year'] )
plt.xlabel( "Pulishing Year" )
plt.ylabel( "Frequency" )
plt.title('Distribution of publishing Year')
plt.show()
```

```python
df['genre'].value_counts().plot(kind="bar")
plt.xlabel("Gener")
plt.ylabel("Number of Books")
plt.title("Number of Books in each genre")
plt.show()
```

```python
df.groupby("Author")['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='genre',y='Book_ratings_count',data=df, hue="genre")
plt.xlabel('Genre')
plt.ylabel('Books Rating Count')
plt.show()
```

```python
plt.scatter(df['sale price'],df['units sold'])
plt.xlabel(' Sales Price ')
plt.ylabel('No of units sold')
plt.title('Scatter plot for sale price vs units sold')
plt.show()
```

```python
#language Distribution of Books


language_counts= df['language_code'].value_counts()
language_counts
```

```python
plt.pie(language_counts,labels=language_counts.index,startangle=90,autopct='%1.1f%%')
plt.title("Language distribution of books")
```

```python
#Publisher revenue list


df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending=False)
```

```python
df.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending=False)
```

```python
df.groupby('Author_Rating')['Book_ratings_count'].median().sort_values(ascending=False)
```
