```python
x=df.drop(columns=['price'],axis=1)
y=df['price']
```

```python
x_one_encode = pd.get_dummies(df,columns=['model','fuelType','transmission'], drop_first=True)
x_one_encode=x_one_encode.astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model','transmission','fuelType']

Xlabel = x.copy()
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    Xlabel[col] = le.fit_transform(Xlabel[col].astype(str))
    label_encoders[col] = le
```

```python
Xlabel.head()
```

```python
from sklearn.preprocessing import StandardScaler

num_cols= ['year','mileage','tax','mpg','engineSize']
scaler=StandardScaler()


x_one_encode[num_cols] = scaler.fit_transform(x_one_encode[num_cols] )
```

```python
Xlabel.columns
```

```python
Xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]= scaler.fit_transform(Xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
Xlabel.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split( x_one_encode, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)

y_pred
```

```python
r2=r2_score(y_test,y_pred)
r2
```

```python
n=X_test.shape[0]
p=X_test.shape[1]
adjR2=1-(((1-r2)*(n-1))/(n-p-1))

adjR2
```
