```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_openml
```

```python
boston = fetch_openml("boston", version=1, as_frame=True,parser='pandas')  

df= boston.frame
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Price']=boston.target
```

```python
df.head()
```

```python
# spliting data into dependent and independent features


x= df.iloc[:,:-1]
y=df['Price']
```

# linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
x=x.astype(float)
y=y.astype(float)
```

```python
model= LinearRegression()
mse=cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=5)

mse
```

```python
mean_mse= np.mean(mse)

mean_mse
```

```python
# separate data into two parts for training and testing

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)

y_pred
```

```python
# compare real and actual values 


compare_df = pd.DataFrame({'Actual':y_test,'Predection':y_pred})
compare_df.head()
```

```python
# model accuracy check R2
r2=r2_score(y_test,y_pred)
r2
```

```python
# adjusted R square 

n=X_test.shape[0]
p=X_test.shape[1]
adjR2=1-(((1-r2)*(n-1))/(n-p-1))

adjR2
```
