```python
df["Profit"] = df["Revenue(INR)"] - df["Budget(INR)"]
```

```python
df.sort_values(by="Budget(INR)", ascending=False)[
    ["Movie_Name", "Budget(INR)", "Revenue(INR)", "Profit"]
].head(10)
```

```python
df[["Budget(INR)", "Revenue(INR)"]].describe()
```

```python
df["ROI"] = df["Revenue(INR)"] / df["Budget(INR)"]
```

```python
df.sort_values(by="Budget(INR)", ascending=False)[
    ["Movie_Name", "Budget(INR)", "Revenue(INR)", "Profit",'ROI']
].head(10)
```

```python
df["Movie_Status"] = df["ROI"].apply(
    lambda x: "Flop" if x < 0.8
    else "Average" if x < 1.2
    else "Hit" if x < 2
    else "Blockbuster"
)
```
